## Line Chart

The line chart is one of the simplest and most widely used data visualization techniques. A line chart displays information as a series of data points or markers connected by straight lines. You can customize the shape, size, color, and other aesthetic elements of the lines and markers for better visual clarity.

Here's a Python list showing the yield of apples (tons per hectare) over six years in an imaginary country called Kanto.

```python
temperature = [90, 92, 98, 95, 100, 102]
```

We can visualize how temperature changes over time using a line chart. To draw a line chart, we can use the `plt.plot` function.

```python
plt.plot(temperature)
```

Calling the `plt.plot` function draws the line chart as expected. 

Let's enhance this plot step-by-step to make it more informative and beautiful.

### Customizing the X-axis

The X-axis of the plot currently shows list element indexes 0 to 5. The plot would be more informative if we could display the month for which we're plotting the data. We can do this by two arguments `plt.plot`.

```python
months_number = [3, 4, 5, 6, 7, 8]
```

```python
plt.plot(months_number, temperature)
```

You can also use text as the x-axis ticks

```python
months_text = ["Mar","Apr","May","Jun","Jul","Aug"]

plt.plot(months_text, temperature)
```

### Axis Labels

We can add labels to the axes to show what each axis represents using the `plt.xlabel` and `plt.ylabel` methods.

```python
plt.plot(months_text, temperature)

plt.xlabel('Months')
plt.ylabel('Temperature')
```

### Plotting Multiple Lines

You can invoke the `plt.plot` function once for each line to plot multiple lines in the same graph. Let's compare the temperature between Tallhassee and NYC.

```python
temp_tlh = [90, 92, 98, 95, 100, 102]
temp_nyc = [88, 84, 91, 95, 96, 93]
```

```python
plt.plot(months_text, temp_tlh)
plt.plot(months_text, temp_nyc)

plt.xlabel('Months')
plt.ylabel('Temperature')
```

### Chart Title and  Legend

To differentiate between multiple lines, we can include a legend within the graph using the `plt.legend` function. We can also set a title for the chart using the `plt.title` function.

```python
plt.plot(months_text, temp_tlh)
plt.plot(months_text, temp_nyc)
plt.xlabel('Months', fontsize=14) #
plt.ylabel('Temperature', fontsize=14) #add fontsize parameter


#Adding a title
plt.title("Monthly Mean Temperature",fontsize = 20)

#Adding a legend
plt.legend(['Tallahassee', 'New York City']);
```

### Line Markers

We can also show markers for the data points on each line using the `marker` argument of `plt.plot`. Matplotlib provides many different markers, like a circle, cross, square, diamond, etc. You can find the full list of marker types here: https://matplotlib.org/stable/api/markers_api.html.

```python
plt.plot(months_text, temp_tlh, marker="o")
plt.plot(months_text, temp_nyc, marker="x")
plt.xlabel('Months')
plt.ylabel('Temperature')

#Adding a title
plt.title("Monthly Mean Temperature")

#Adding a legend
plt.legend(['Tallahassee', 'New York City']);
```

### Styling Lines and Markers

The `plt.plot` function supports many arguments for styling lines and markers:

- `color` or `c`: Set the color of the line ([supported colors](https://matplotlib.org/stable/gallery/color/named_colors.html))
- `linestyle` or `ls`: Choose between a solid or dashed line
- `linewidth` or `lw`: Set the width of a line
- `markersize` or `ms`: Set the size of markers
- `markeredgecolor` or `mec`: Set the edge color for markers
- `markeredgewidth` or `mew`: Set the edge width for markers
- `markerfacecolor` or `mfc`: Set the fill color for markers
- `alpha`: Opacity of the plot


Check out the documentation for `plt.plot` to learn more: [https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) .

```python
plt.plot(months_text, temp_tlh, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(months_text, temp_nyc, marker='o', c='r', ls='--', lw=3, ms=10, alpha=0.5)


plt.xlabel('Months')
plt.ylabel('Temperature')

#Adding a title
plt.title("Monthly Mean Temperature")

#Adding a legend
plt.legend(['Tallahassee', 'New York City']);
```

### Changing the Figure Size

You can use the `plt.figure` function to change the size of the figure.

```python
plt.figure(figsize=(12, 6)) #12 by 6


plt.plot(months_text, temp_tlh, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(months_text, temp_nyc, marker='o', c='r', ls='--', lw=3, ms=10, alpha=0.5)

plt.xlabel('Months')
plt.ylabel('Temperature')

#Adding a title
plt.title("Monthly Mean Temperature")

#Adding a legend
plt.legend(['Tallahassee', 'New York City']);
```

## Scatter Plot

In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid. Additionally, you can also use a third variable to determine the size or color of the points. Let's try out an example.

Here we use a simple taxis dataset from the seaborn library as one of the example datasets.

```python
# Load data into a Pandas dataframe
taxi = sns.load_dataset("taxis")
```

```python
taxi.head()
```

Let's try to visualize the relationship between taxi trip distance and trip total cost. 

You can use `plt.scatter` or `sns.scatterplot` to make scatterplots

```python
plt.scatter(taxi.distance, taxi.total)
```

```python
sns.scatterplot(x=taxi.distance, y=taxi.total)
```

For both plots you can use the `color` parameter to control the color and `s` parameter to control the size of the dot.

```python
sns.scatterplot(x=taxi.distance, y=taxi.total,color='red',s=10)
```

### Plotting using Pandas Data Frames

Seaborn has in-built support for Pandas data frames. Instead of passing each column as a series, you can provide column names and use the `data` argument to specify a data frame.

```python
plt.title('Taxis Trip Distance vs. Cost',fontsize=14)

sns.scatterplot(x='distance', 
                y='total', 
                s=50,
                data=taxi);
```

## Histogram

A histogram represents the distribution of a variable by creating bins (interval) along the range of values and showing vertical bars to indicate the number of observations in each bin. 

For example, let's visualize the distribution of values of distance the taxi dataset. We can use the `plt.hist` function to create a histogram.

```python
taxi.distance
```

```python
plt.title("Distribution of Taxi Trip Distances")

plt.hist(taxi.distance)
```

We can see that the most trip distance is short, say, below 10 miles.

### Controlling the size and number of bins

We can control the number of bins or the size of each one using the bins argument.


```python
plt.hist(taxi.distance, bins=20)
```

```python
import numpy as np

# Specifying the boundaries of each bin
plt.hist(taxi.distance, bins=np.arange(0, 30, 2.5));
```

### Multiple Histograms

Similar to line charts, we can draw multiple histograms in a single chart. We can reduce each histogram's opacity  so that one histogram's bars don't hide the others'.

Let's draw separate histograms for each pick up loctaion in NYC.

```python
taxi.pickup_borough.unique()
```

```python
#manh_df = taxi[taxi.pickup_borough == 'Manhattan']

quee_df = taxi[taxi.pickup_borough == 'Queens']

bron_df = taxi[taxi.pickup_borough == 'Bronx']

broo_df = taxi[taxi.pickup_borough == 'Brooklyn']
```

```python
bins = np.arange(0, 30, 2.5)
#plt.hist(manh_df.distance, alpha=0.4, )
plt.hist(quee_df.distance, alpha=0.4, bins=bins)
plt.hist(bron_df.distance, alpha=0.4, bins=bins)
plt.hist(broo_df.distance, alpha=0.4, bins=bins)

plt.title('Distribution of Taxi Trip by Pick up Location')
plt.legend(['Queens', 'Bronx', 'Brooklyn']);
```

```python
plt.bar(months_text, temp_tlh)

plt.xlabel("Temperature")
```

### Bar Plots with Averages and Confindece intervals/variances

Let's look at another sample dataset included with Seaborn, called `tips`. The dataset contains information about the sex, time of day, total bill, and tip amount for customers visiting a restaurant over a week.

```python
tips_df = sns.load_dataset("tips");
```

```python
tips_df
```

We might want to draw a bar chart to visualize how the average bill amount varies across different days of the week. One way to do this would be to compute the day-wise averages and then use `plt.bar` (try it as an exercise).

However, since this is a very common use case, the Seaborn library provides a `barplot` function which can automatically compute averages.

```python
sns.barplot(x='day', y='total_bill', data=tips_df,)
```

The lines cutting each bar represent the amount of variation (95% confidencei nterval) in the values. For instance, it seems like the variation in the total bill is relatively high on Fridays and low on Saturday.

We can also specify a `hue` argument to compare bar plots side-by-side based on a third feature, e.g., sex.

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, errorbar = None)
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df)
```

You can make the bars horizontal simply by switching the axes.

```python
sns.barplot(x='total_bill', y='day', hue='sex', data=tips_df)

plt.xlim(0, 40) # To provide a range for x or y axis
```

## Heatmap

A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors. The best way to understand it is by looking at an example. We'll use another sample dataset from Seaborn, called `flights`, to visualize monthly passenger footfall at an airport over 12 years.

```python
flights_raw = sns.load_dataset("flights")
```

```python
flights_raw.head()
```

Note: check what the `.pivot` function is doing here.

```python
flights_df = flights_raw.pivot(index="month", columns="year", values="passengers")
```

```python
flights_df
```

`flights_df` is a matrix with one row for each month and one column for each year. The values show the number of passengers (in thousands) that visited the airport in a specific month of a year. We can use the `sns.heatmap` function to visualize the footfall at the airport.

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df)
```

The brighter colors indicate a higher footfall at the airport. By looking at the graph, we can infer two things:

- The footfall at the airport in any given year tends to be the highest around July & August.
- The footfall at the airport in any given month tends to grow year by year.

We can also display the actual values in each block by specifying `annot=True` and using the `cmap` argument to change the color palette.

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df, fmt="d", annot=True, cmap='Blues')
```

## Plotting multiple charts in a grid

Matplotlib and Seaborn also support plotting multiple charts in a grid, using `plt.subplots`, which returns a set of axes for plotting. 

Here's a single grid showing the different types of charts we've covered in this tutorial.

Finally, you can save the figure as a static image using `plt.savefig` function. The `dpi` in the function controls the resolution of your image. Usually >300 is a good resolution.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8)) # 2 by 2 layout with size 12 by 8


# A matplotlib plot
axes[0,0].plot(months_text, temp_tlh, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
axes[0,0].plot(months_text, temp_nyc, marker='o', c='r', ls='--', lw=3, ms=10, alpha=0.5)
axes[0,0].set_xlabel('Months')
axes[0,0].set_ylabel('Temperature');
axes[0,0].set_title("Monthly Mean Temperature")
axes[0,0].legend(['Tallahassee', 'New York City']);


# A seaborn plot
sns.barplot(ax=axes[0,1], x='total_bill', y='day', hue='sex', data=tips_df)


# A matplotlib plot
axes[1,0].bar(months_text, temp_tlh)
axes[1,0].set_xlabel("Temperature")


# A seaborn plot
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1])



plt.tight_layout(pad=2) #add a small padding



#Save the figure to a local image file
#plt.savefig("my_plots.png",dpi=300)
```
