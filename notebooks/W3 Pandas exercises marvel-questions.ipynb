### Create a `marvel_df` pandas DataFrame with the given marvel data


```python
marvel_data = [
    ['Spider-Man', 'male', 1962],
    ['Captain America', 'male', 1941],
    ['Wolverine', 'male', 1974],
    ['Iron Man', 'male', 1963],
    ['Thor', 'male', 1963],
    ['Thing', 'male', 1961],
    ['Mister Fantastic', 'male', 1961],
    ['Hulk', 'male', 1962],
    ['Beast', 'male', 1963],
    ['Invisible Woman', 'female', 1961],
    ['Storm', 'female', 1975],
    ['Namor', 'male', 1939],
    ['Hawkeye', 'male', 1964],
    ['Daredevil', 'male', 1964],
    ['Doctor Strange', 'male', 1963],
    ['Hank Pym', 'male', 1962],
    ['Scarlet Witch', 'female', 1964],
    ['Wasp', 'female', 1963],
    ['Black Widow', 'female', 1964],
    ['Vision', 'male', 1968]
]
```

```python
marvel_df = pd.DataFrame(marvel_data)
marvel_df
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Add column names ['name', 'sex', 'first_appearance'] to the `marvel_df`
 

```python
marvel_df.columns = ['name', 'sex', 'first_appearance']
marvel_df
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Drop 'Namor' and 'Hank Pym' rows


```python
condition = ~marvel_df.name.isin(["Namor", "Hank Pym"])
marvel_df[condition]
```

## DataFrame manipulation and operations

### Modify the `first_appearance` of 'Vision' to year 1964

```python
marvel_df.loc[marvel_df["name"] == "Vision", "first_appearance"] = 1964
marvel_df.tail()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Add a new column to `marvel_df` called 'years_since' with the years since `first_appearance` til now 2024


```python
marvel_df["years_since"] = 2024 - marvel_df["first_appearance"]
marvel_df
```

## DataFrame boolean arrays (also called masks)

### Given the `marvel_df` pandas DataFrame, show all the female characters


```python
fonly = marvel_df["sex"] == "female"
marvel_df[fonly]
```

### Given the `marvel_df` pandas DataFrame, get all male characters


```python
monly = marvel_df["sex"] == "male"
marvel_df[monly]
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Given the `marvel_df` pandas DataFrame, get the characters with `first_appearance` after 1970
