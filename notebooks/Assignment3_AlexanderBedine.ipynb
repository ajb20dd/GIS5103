## Cleaning & Merging Datasets

```python
voting_2020.columns = ["FIPS Code", "Percent of Dem Voters"]
voting_2020
```

rename columns of voting_2020 data to match with Education data, using pandas' [columns](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.columns.html) function for dataframes.

```python
merged = pd.merge(voting_2020, education, on="FIPS Code")
merged
```

merge dataframes on "FIPS Code" column since that's the common ground using pandas' [merge](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html) function.

```python
merged = merged.iloc[:, [0,1,7,2,3,4,5,6]]
merged
```

Reordering columns to make data more readable.

```python
merged.sort_values('FIPS Code', ascending=True)
```

Sort the data by FIPS Code.

```python
merged.isna().sum()
```

removing indexes with null data to improve analysis

```python
cleanmerge = merged.dropna()
```

```python
voter_education = cleanmerge.drop(["Percent of adults with less than a high school diploma, 2018-22", "Percent of adults with a high school diploma only, 2018-22", "Percent of adults completing some college or associate's degree, 2018-22"], axis=1)
```
