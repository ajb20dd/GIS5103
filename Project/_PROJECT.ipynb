So we need to group these by county

```python
agg_FL_SSL = FL_SSL_dtype.groupby('county_fips').agg({
    'Total_Pb': 'sum', 
    'Total': 'sum'
}).reset_index()
```

```python
agg_FL_SSL
```

```python
agg_FL_SSL['pct_Pb'] = (agg_FL_SSL['Total_Pb'] / agg_FL_SSL['Total']).round(3)
agg_FL_SSL
```

Now we have the percentage of lead pipe infrastructure in each county (hopefully)

```python
agg_FL_SSL.dtypes
```

```python
FL_SSL_county = agg_FL_SSL.merge(FL_counties, left_on='county_fips', right_on='COUNTY')
FL_SSL_county
```
